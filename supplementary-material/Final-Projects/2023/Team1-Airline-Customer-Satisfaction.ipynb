## Motivation
Satisfaction rate can help airlines to assess the gaps to provide better service and customer experience. The exploratary analysis of this data can help narrow the scope of work for the airlines by identifying major problem areas.


Airlines that face low customer satisfaction encounters several challenges that can impact their business and operations. 
Some of the challenges airlines may face due to low customer satisfaction include:

**Decreased customer loyalty:** Low customer satisfaction can result in decreased customer loyalty, as dissatisfied customers may be less likely to choose the airline for future travel or recommend it to others. This can lead to decreased repeat business and reduced customer retention, which can negatively impact the airline's revenue and profitability.

**Negative word-of-mouth and reputation damage**: Dissatisfied customers are more likely to share their negative experiences with others through word-of-mouth, online reviews, and social media, which can damage the airline's reputation. Negative publicity can spread quickly and impact the airline's brand image, resulting in potential loss of new customers and reduced market share.

**Increased customer complaints and service recovery costs:** Low customer satisfaction may result in an increased number of customer complaints, which can require additional resources and efforts to address and resolve. Service recovery costs, including compensations, refunds, or other remedies offered to dissatisfied customers, can also add to the operational costs of the airline.

**Reduced revenue and profitability:** Satisfied customers are more likely to be loyal and make repeat purchases, which can contribute to the airline's revenue and profitability. Conversely, low customer satisfaction can result in decreased revenue due to reduced bookings, lower ticket prices to attract customers, and increased costs associated with addressing customer complaints and issues.

**Difficulty in attracting new customers:** Airlines with low customer satisfaction may struggle to attract new customers, as negative reviews and poor reputation can deter potential customers from choosing the airline. This can result in decreased market share and limited opportunities for growth.

**Increased competition:** In today's competitive airline industry, customer satisfaction plays a crucial role in attracting and retaining customers. Airlines with low customer satisfaction may face increased competition from other airlines that provide better customer experiences, leading to further challenges in attracting and retaining customers.

**Operational inefficiencies:** Low customer satisfaction may indicate potential operational issues or gaps in service quality that need to be addressed. Addressing these operational inefficiencies may require additional resources, training, and process improvements, which can pose challenges for airlines in terms of time, effort, and costs.

###PreProcessing- Handling Categorical Data

We are starting with Logistic Regression to create a prediction model and since our data contains both numberical and categorical values, we are going to process our categorical data first. 

```python

aa_df2 = df
```

**Finding categorical variables** that exist in the data and their unique values are-
1. satisfaction - 'satisfied', 'dissatisfied'
2. Gender - 'Female','Male'
3. Customer Type -'Loyal Customer', 'disloyal Customer'
4. Type of Travel - 'Personal Travel', 'Business travel'
5. Class - 'Eco', 'Business', 'Eco Plus'

```python
# loop through each column and check if it's categorical
for column in aa_df2.columns:
    if aa_df2[column].dtype == 'object':
        print(f"{column} has categories: {aa_df2[column].unique()}")
```


**Converting binary categorical variables** to numerical binary variables

```python
# convert binary categorical variables to numerical binary variables
aa_df2['Gender'] = aa_df2['Gender'].map({'Male': 1, 'Female': 0})
aa_df2['Customer Type'] = aa_df2['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 0})
aa_df2['Type of Travel'] = aa_df2['Type of Travel'].map({'Personal Travel': 1, 'Business travel': 0})
aa_df2['satisfaction'] = aa_df2['satisfaction'].map({'satisfied': 1, 'dissatisfied': 0})
```

**One Hot Encoding**
The last categorical variable, Class, has three categories, Business, Eco and Eco Plus. One Hot Encoder is used to be able to apply regression models on the dataset.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# create an instance of OneHotEncoder with the 'Class' column

ohe = OneHotEncoder(sparse=True, sparse_output=False)
ct = ColumnTransformer([('ohe', ohe, ['Class'])])

# fit_transform the data
X_encoded = ct.fit_transform(aa_df2)

# get the names of the one-hot encoded features
feature_names = ct.named_transformers_['ohe'].get_feature_names_out(['Class'])
```

```python
from sklearn.preprocessing import OneHotEncoder

# Create an instance of the OneHotEncoder class
encoder = OneHotEncoder(sparse=False)

# Fit and transform the 'Class' variable
class_encoded = encoder.fit_transform(aa_df2[['Class']])


# Create a new dataframe with the one-hot encoded 'Class' variable
class_aa_df2 = pd.DataFrame(class_encoded, columns=feature_names)

# Concatenate the one-hot encoded 'Class' variable with the original dataframe
aa_df2_encoded = pd.concat([aa_df2, class_aa_df2], axis=1)

# Drop the original 'Class' variable
aa_df2_encoded.drop('Class', axis=1, inplace=True)
```

###Running Regression - Logistic, Lasso, Ridge

**Logistic Regression** Running the model after splitting the dataset into 70 and 30 using random value 833. Obtained **accuracy of 83.5%**

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Create the feature matrix X by dropping the target variable and any other irrelevant variables
X = aa_df2_encoded.drop(['satisfaction'], axis=1)

# Create the target vector y
y = aa_df2_encoded['satisfaction']

# Split the data into train and test sets with a 70-30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=833)

# Initialize the logistic regression model
logreg = LogisticRegression(max_iter=3000)

# Train the model on the training data
logreg.fit(X_train, y_train)

# Predict the labels of the test data
y_pred = logreg.predict(X_test)

# Evaluate the performance of the model on the test data
accuracy = logreg.score(X_test, y_test)
print("Accuracy:", accuracy*100)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# get the predicted probabilities and calculate the false positive rate (fpr) and true positive rate (tpr)
y_prob = logreg.predict_proba(X_test)[:, 1]
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# calculate the area under the curve (AUC) and plot the ROC curve
roc_auc = auc(fpr, tpr)
plt.plot(fpr, tpr, label='ROC curve (AUC = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```

By using **Lasso Regression**, we obtained **accuracy of 79.4%**

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create the feature matrix X by dropping the target variable and any other irrelevant variables
X = aa_df2_encoded.drop(['satisfaction'], axis=1)

# Create the target vector y
y = aa_df2_encoded['satisfaction']

# Split the data into train and test sets with a 70-30 split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=833)

# Perform Lasso regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)

# Get the coefficients and feature names
coef = lasso.coef_
feature_names = aa_df2_encoded.drop('satisfaction', axis=1).columns

# Get the top 5 most relevant features
top_features = sorted(zip(coef, feature_names), reverse=True)[:5]
print("Top 5 most relevant features:")
for feat in top_features:
    print(feat)

# Get the bottom 5 least relevant features
bottom_features = sorted(zip(coef, feature_names))[:5]
print("\nBottom 5 least relevant features:")
for feat in bottom_features:
    print(feat)

# Predict on the test set and calculate accuracy
y_pred = lasso.predict(X_test)
accuracy = accuracy_score(y_test, y_pred.round())
print("\nAccuracy:", accuracy*100)
```
